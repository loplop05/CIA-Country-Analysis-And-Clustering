```python
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

```python
mask = df['Literacy (%)'].isna()
```

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(
    df.groupby('Region')['Literacy (%)'].transform('mean')
)
```

```python
""" to see the filled value for each Region """

region_mean = df.groupby('Region')['Literacy (%)'].transform('mean')

df.loc[mask, ['Region']].assign(Filled_Value=region_mean[mask])
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df = df.dropna()
df.isna().sum().sum()
```

```python
X = df.drop("Country",axis=1)
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(X)
```

```python
X_Scaled
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = []
for k in range(2,30):
    model = KMeans(n_clusters=k)
    model.fit(X_Scaled)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
```
